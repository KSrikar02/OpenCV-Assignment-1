```python
import cv2
import numpy as np

# Define the piecewise linear transformation function
def piecewise_linear(x):
    if x < 50:
        return 0.5*x
    elif x < 200:
        return 5*x + 225
    else:
        return 0.5*x + 160

# Read in the video file
cap = cv2.VideoCapture(r"C:\Users\umasr\Downloads\DIP assignment\Q3\moon_gray_contrast.mp4")

# Get the number of frames in the video
num_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

# Get the frame rate of the video
fps = int(cap.get(cv2.CAP_PROP_FPS))

# Get the size of the frames in the video
frame_size = (int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)), int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)))

# Define the codec and create a VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('output.mp4', fourcc, fps, frame_size, isColor=False)

# Loop through each frame of the video
while True:
    # Read in the current frame
    ret, frame = cap.read()

    # Check if the frame was successfully read in
    if not ret:
        break

    # Convert the frame to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Apply the piecewise linear transformation to the frame
    lut = np.vectorize(piecewise_linear)
    transformed = lut(gray)

    # Convert the data type of the transformed image(uint8 format)
    transformed = cv2.convertScaleAbs(transformed)

    # Write the transformed frame to the output video file
    out.write(transformed)

    # Display the transformed frame
    cv2.imshow('Transformed Frame', transformed)

    # Wait for a key press
    key = cv2.waitKey(1) & 0xFF

    # If the key pressed is q, break out of the loop
    if key == ord('q'):
        break

# Release the video file, the VideoWriter object, and destroy all windows
cap.release()
out.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# Load the video
cap = cv2.VideoCapture(r"C:\Users\umasr\Downloads\DIP assignment\Q3\moon_gray_contrast.mp4")

# Check if the video was opened successfully
if not cap.isOpened():
    print("Error opening video file")

# Get the frames per second (fps) and frame size
fps = int(cap.get(cv2.CAP_PROP_FPS))
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Define the Sobel kernel
kernel = np.array([[-1, -4, -6, -4, -1],
                   [-2, -8, -12, -8, -2],
                   [0, 0, 0, 0, 0],
                   [2, 8, 12, 8, 2],
                   [1, 4, 6, 4, 1]])

# Define the sharpening filter
sharpen_kernel = np.array([[-1, -1, -1],
                           [-1,  9, -1],
                           [-1, -1, -1]])

# Define the output video file
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('filtered_video.mp4', fourcc, fps, (width, height), isColor=False)

# Loop through each frame of the video
while cap.isOpened():
    # Read the frame
    ret, frame = cap.read()

    # If the frame was read successfully
    if ret:
        # Convert the frame to grayscale
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

        # Apply the Sobel kernel to the grayscale frame
        sobel = cv2.filter2D(gray, -1, kernel)

        # Add the Sobel result to the grayscale frame
        sharpened = cv2.addWeighted(gray, 1, sobel, 0.5, 0)

        # Apply the sharpening filter to the result
        result = cv2.filter2D(sharpened, -1, sharpen_kernel)

        # Write the result to the output video file
        out.write(result)

        # Display the result
        cv2.imshow('result', result)

        # Exit if the 'q' key is pressed
        if cv2.waitKey(25) & 0xFF == ord('q'):
            break
    # If the frame could not be read
    else:
        break

# Release the resources
cap.release()
out.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# Load video
cap = cv2.VideoCapture(r"C:\Users\umasr\Downloads\DIP assignment\Q3\moon_gray_contrast.mp4")

# Check if video file is opened successfully
if (cap.isOpened() == False): 
print("Error opening video stream or file")

# Get video frame dimensions
frame_width = int(cap.get(3))
frame_height = int(cap.get(4))

# Define output video codec and filename
codec = cv2.VideoWriter_fourcc(*'mp4v') # choose the desired codec
filename = 'sliced_video.mp4'

# Define output video writer
out = cv2.VideoWriter(filename, codec, 30.0, (frame_width,frame_height), isColor=False)

# Iterate through frames
while (cap.isOpened()):
    # Read frame
    ret, frame = cap.read()

    # Check if frame is read correctly
    if ret == True:
        # Convert frame to grayscale
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

        # Initialize empty array for sliced frames
        sliced_frames = np.zeros_like(gray, dtype=np.uint8)

        # Loop through bits 0-7 and slice them
        for i in range(8):
            bit_plane = np.zeros_like(gray, dtype=np.uint8)
            bit_plane = np.bitwise_and(gray, 2**i)
            bit_plane = np.right_shift(bit_plane, i)
            sliced_frames += bit_plane

        # Write sliced frame to output video
        out.write(sliced_frames)

        # Display sliced frame
        cv2.imshow('Sliced frames', sliced_frames)

        # Exit if 'q' is pressed
        if cv2.waitKey(25) & 0xFF == ord('q'):
            break
    else:
        break

# Release video capture, video writer, and close window
cap.release()
out.release()
cv2.destroyAllWindows()
```
